```python
%matplotlib inline
import numpy as np
import cv2 
def average_filter(image, ksize): 
    rows, cols = image.shape[:2] 
    dst = np.zeros((rows, cols), np.uint8) 
    center = ksize // 2 
    for i in range(rows): 
        for j in range(cols): 
            y1, y2 = i - center, i + center + 1 
            x1, x2 = j - center, j + center + 1 
            if y1 < 0 or y2 > rows or x1 < 0 or x2 > cols: 
                dst[i, j] = image[i, j] 
            else: 
                mask = image[y1:y2, x1:x2] 
                dst[i, j] = cv2.mean(mask)[0] 
    return dst 
image = cv2.imread('/home/inlab/anaconda3/Alien.png',cv2.IMREAD_GRAYSCALE) 
print(image.shape)
if image is None: raise Exception("영상파일 읽기 오류") 

avg_img = average_filter(image, 5) 
kernelSizes = [(3, 3), (9, 9), (15, 15)]
for (kX, kY) in kernelSizes:
    blur_img = cv2.blur(image, (kX, kY)) 
box_img = cv2.boxFilter(image, ddepth=-1, ksize=(5, 5) ) 

cv2.imshow("image", image) 
cv2.imshow("avg_img", avg_img) 
cv2.imshow("blur_img", blur_img) 
cv2.imshow("box_img", box_img) 
cv2.waitKey(0)
```
